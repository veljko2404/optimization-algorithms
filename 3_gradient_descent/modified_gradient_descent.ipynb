```python
import numpy as np
import time

def rosenbrock(x):
    x = np.asarray(x, float)
    return np.sum(100.0 * (x[1:] - x[:-1]**2)**2 + (1.0 - x[:-1])**2)

def rosenbrock_grad(x):
    x = np.asarray(x, float)
    g = np.zeros_like(x)
    g[0] = -400*x[0]*(x[1] - x[0]**2) - 2*(1 - x[0])
    g[1:-1] = 200*(x[1:-1] - x[:-2]**2) - 400*x[1:-1]*(x[2:] - x[1:-1]**2) - 2*(1 - x[1:-1])
    g[-1] = 200*(x[-1] - x[-2]**2)
    return g

def modified_gradient_descent(f, grad_f, x0, step=1e-3, tol=1e-6, max_iter=200000, beta=0.5):
    x = np.asarray(x0, float).copy()
    t0 = time.process_time()
    k = 0
    while k < max_iter and np.linalg.norm(grad_f(x)) > tol:
        d = -grad_f(x)
        a = step
        fx = f(x)
        while f(x + a*d) >= fx:
            a *= beta
            if a < 1e-16:
                break
        x = x + a*d
        k += 1
    t1 = time.process_time()
    print(f"CPU time: {t1 - t0:.6f} s")
    return x, f(x), k

n = 2
x0 = np.full(n, -1.2, float)
x0[1::2] = 1.0

x_star, f_star, iters = modified_gradient_descent(rosenbrock, rosenbrock_grad, x0, step=1e-3)
print("Final point:", x_star)
print("Iterations:", iters)
print("Function value:", f_star)
```
