```python
import numpy as np
from plot_function import plot_function


def dichotomy_method(f, a, b, delta_min, eps=1e-5, max_iter=10000):
    low = float(a)
    high = float(b)
    path = []

    it = 0
    x_best = (low + high) / 2.0
    f_best = f(x_best)

    while (high - low) > delta_min and it < max_iter:
        it += 1

        mid = (low + high) / 2.0
        x1 = mid - eps
        x2 = mid + eps

        f1 = f(x1)
        f2 = f(x2)

        if f1 < f2:
            high = x2
            x_best, f_best = x1, f1
        else:
            low = x1
            x_best, f_best = x2, f2

        path.append(x_best)

    return x_best, f_best, path


def f(x):
    return x**4 - 4*x**3 - 6*x**2 + 24*x + 5


a, b = -3.0, 6.0
x_min, f_min, path = dichotomy_method(f, a, b, delta_min=0.001, eps=1e-5)

xs = np.linspace(a, b, 1000)
ys = f(xs)

print(f"x_min = {x_min:.4f}")
print(f"f(x_min) = {f_min:.6f}")

plot_function(xs, ys, path, f, "x**4 - 4*x**3 - 6*x**2 + 24*x + 5", x_min, "Dichotomous Search (Dichotomy) Method")
```
