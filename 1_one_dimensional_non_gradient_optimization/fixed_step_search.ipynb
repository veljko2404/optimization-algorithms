```python
from plot_function import plot_function
import numpy as np

def fixed_step_scan(f, a, b, delta=0.1):
    x = a
    x_min = a
    f_min = f(a)
    path = [x]

    while x <= b:
        x += delta
        if f(x) < f_min:
            f_min = f(x)
            x_min = x
        path.append(x)
    return x_min, path


def f(x):
    return x**4 - 4*x**3 - 6*x**2 + 24*x+5


x_min, path = fixed_step_scan(f, -3, 5, 0.1)
xs = np.linspace(-3.5, 6, 400)
ys = f(xs)

print(f"x_min = {x_min:.4f}")
print(f"f(x_min) = {f(x_min):.6f}")

plot_function(xs, ys, path, f, "x**4 - 4*x**3 - 6*x**2 + 24*x+5", x_min, "Fixed step search")
```
