```python
from plot_function import plot_function
import numpy as np


def adaptive_step_scn(f, a, b, delta=1.0, delta_min=0.01):
    f_min = f(a)
    x_min = a
    L = 4

    path = [a]
    while delta >= delta_min:
        x = a
        while x < b:
            x += delta
            path.append(x)
            if x > b:
                break

            if f(x) < f_min:
                f_min = f(x)
                x_min = x
        a = x_min - delta
        b = x_min + delta
        delta /= L
    return x_min, path


def f(x):
    return x**4 - 4*x**3 - 6*x**2 + 24*x+5


x_min, path = adaptive_step_scn(f, -3, 6, 0.5, 0.001)
xs = np.linspace(-3.5, 6, 500)
ys = f(xs)

print(f"x_min = {x_min:.4f}")
print(f"f(x_min) = {f(x_min):.6f}")

plot_function(xs, ys, path, f, "x**4 - 4*x**3 - 6*x**2 + 24*x + 5", x_min, "Adaptive Step Search")
```
