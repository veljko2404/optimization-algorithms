```python
import numpy as np
from plot_function import plot_function


def one_d_simplex_method(f, x0, delta, delta_min, max_iter):
    f_min = f(x0)
    x_min = x0
    x = x0 + delta
    f_pr = f(x0)
    id_flag = 0
    L = 4
    path = [x0, x]
    it = 0
    if f(x) > f(x0):
        delta = -delta
        x = x0 + delta
        path[-1] = x

    while abs(delta) >= delta_min and it < max_iter:
        it += 1
        fx = f(x)
        if fx < f_pr:
            if fx < f_min:
                f_min = fx
                x_min = x

            if id_flag == 0:
                delta = 2 * delta
        else:
            id_flag = 1
            delta = -delta / L

        f_pr = fx
        x = x + delta
        path.append(x)

    return x_min, f_min, path


def f(x):
    return x ** 4 - 4 * x ** 3 - 6 * x ** 2 + 24 * x + 5

x_min, f_min, path = one_d_simplex_method(f, -3, 1.0, 0.01, 10000)
xs = np.linspace(-3.5, 6.5, 500)
ys = f(xs)

print(f"x_min = {x_min:.4f}")
print(f"f(x_min) = {f(x_min):.6f}")
plot_function(xs, ys, path, f, "x**4 - 4*x**3 - 6*x**2 + 24*x + 5", x_min, "1D Simplex Method v2")
```
