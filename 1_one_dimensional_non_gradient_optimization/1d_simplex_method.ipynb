```python
import numpy as np

from plot_function import plot_function


def one_d_simplex_method(f, a, delta=1.0, delta_min=0.01):
    f_min = f(a)
    x_min = a
    x = a
    f_pr = f(a)
    L = 4
    path = [x]
    while abs(delta) >= delta_min:
        x += delta
        path.append(x)

        if f(x) < f_min:
            f_min = f(x)
            x_min = x

        if f(x) > f_pr:
            delta = -delta / L

        f_pr = f(x)

    return x_min, path


def f(x):
    return x**4 - 4*x**3 - 6*x**2 + 24*x+5


x_min, path = one_d_simplex_method(f, -3, 1.0, 0.01)
xs = np.linspace(-3.5, 5, 500)
ys = f(xs)

print(f"x_min = {x_min:.4f}")
print(f"f(x_min) = {f(x_min):.6f}")

plot_function(xs, ys, path, f, "x**4 - 4*x**3 - 6*x**2 + 24*x + 5", x_min, "1D Simplex Method")
```
