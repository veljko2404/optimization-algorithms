```python
from plot_function import plot_function
import numpy as np


def golden_section_method(f, a, b, tol=1e-4, max_iter=10000):
    phi = (np.sqrt(5) - 1) / 2
    low = float(a)
    high = float(b)

    x1 = high - phi * (high - low)
    x2 = low + phi * (high - low)

    f1 = f(x1)
    f2 = f(x2)

    path = []

    it = 0
    while (high - low) > tol and it < max_iter:
        it += 1

        if f1 < f2:
            high = x2
            x2 = x1
            f2 = f1

            x1 = high - phi * (high - low)
            f1 = f(x1)
            x_best, f_best = x1, f1
        else:
            low = x1
            x1 = x2
            f1 = f2

            x2 = low + phi * (high - low)
            f2 = f(x2)
            x_best, f_best = x2, f2

        path.append(x_best)

    return x_best, f_best, path

def f(x):
    return x**4 - 4*x**3 - 6*x**2 + 24*x + 5


x_min, f_min, path = golden_section_method(f, -3.0, 6.0)
xs = np.linspace(-3, 6.0, 1000)
ys = f(xs)

print(f"x_min = {x_min:.4f}")
print(f"f(x_min) = {f_min:.6f}")
plot_function(xs, ys, path, f, "x**4 - 4*x**3 - 6*x**2 + 24*x + 5", x_min, "Golden Section Search")
```
